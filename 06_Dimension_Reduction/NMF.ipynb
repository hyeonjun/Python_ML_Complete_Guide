# NMF(Non-Negative Matrix Factorization)

* Truncated SVD와 같이 낮은 랭크를 통한 행렬 근사(Now-Rank Approximation) 방식의 변형이다.
* 원본 행렬 내의 모든 원소 값이 모두 양수(0 이상)이면 두 개의 양수 행렬로 분해될 수 있는 기법

    image.png

4 * 6 원본 행렬 V는 4 * 2 행렬 W와 2 * 6 행렬 H로 근사해 분해될 수 있다.

* 행렬 분해(Matrix Factorization)는 SVD와 같은 행렬 분해 기법을 통칭하며, 행렬 분해를 하면 길고 가는 행렬 W, 작고 넓은 H로 분해된다.
* W 행렬 : 원본 행렬과 같은 행 크기. 원본 행에 대해서 잠재 요소 값이 얼마나 되는지
* H 행렬 : 원본 행렬의 같은 열 크기. 잠재 요소가 원본 열(원본 속성)로 어떻게 구성되었는지
* 이미지 변환 및 압축, 텍스트의 토픽 도출 등 영역에서 사용
* 사이킷런의 NMF 클래스

```python
from sklearn.decomposition import NMF
from sklearn.datasets import load_iris
import matplotlib.pyplot as plt
%matplotlib inline

iris = load_iris()
iris_ftrs = iris.data
nmf = NMF(n_components=2)
nmf.fit(iris_ftrs)
iris_nmf = nmf.transform(iris_ftrs)
plt.scatter(x=iris_nmf[:, 0], y=iris_nmf[:, 1], c=iris.target)
plt.xlabel('NMF Component 1')
plt.ylabel('NMF Component 2')
```

NMF도 SVD와 유사하게 이미지 압축을 통한 패턴 인식, 텍스트의 토픽 모델링 기법, 문서 유사도 및 클러스러링에 사용된다. 또한 영화 추천과 같은 추천(Recommendations) 영역에 활발하게 적용된다.

사용자의 상품(예 : 영화) 평가 데이터 세트인 사용자-평가 순위(user-Rating) 데이터 세트를 행렬 분해 기법을 통해 분해하면서 사용자가 평가하지 않은 상품에 대한 잠재적인 요소를 추출해 이를 통해 평가 순위(Rating)를 예측하고, 높은 순위로 예측된 상품을 추천해주는 방식(이를 잠재 요소(Latent Ractoring) 기반의 추천 방식).

# 정리

* 대표적인 차원 축소 알고리즘 = PCA, LDA, SVD, NMF

* 차원 축소 : 많은 피처로 이루어진 데이터 세트를 차원 축소를 통해 더욱 직관적으로 이해할 수 있게 함.

* 차원 축소는 단순히 피처 개수를 줄이는 것이 아닌, 잠재적 요소를 추출하는 것.
<br/>
---
<br/>

* PCA는 입력 데이터의 변동성이 가장 큰 축을 구하고, 다시 이 축에 직각인 축을 반복적으로 축소하려는 차원 개수만큼 구한 뒤 입력 데이터를 이 축들에 투영해 차원 축소 함.

* PCA는 차원 축소를 위해 입력 데이터의 공분산 행렬을 기반으로 고유 벡터를 생성. 이 고유 벡터에 입력 데이터를 선형 변환하는 방식.
<br/>
---
<br/>

* LDA는 입력 데이터의 결정 값 클래스를 최대한 분리할 수 있는 축을 찾는 방식으로 차원 축소
<br/>
---
<br/>

* SVD는 매우 많은 피처를 가진 고차원 행렬을 두 개의 저차원 행렬로 분리하는 행렬 분해 기법.

* SVD는 행렬 분해를 수행하면서 원본 행렬에서 잠재된 요소를 추출하기 때문에 토픽 모델링이나 추천 시스템에서 사용된다.
